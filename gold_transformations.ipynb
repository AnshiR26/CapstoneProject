```python
import dlt
from pyspark.sql import functions as F
from pyspark.sql.functions import col, when, concat_ws
```

```python
@dlt.table(    
           name="gold_bank_churn",    
           comment="Bank churn merged with CustomerInfo using CustomerID",    
           table_properties={"quality": "silver"}
)
def gold_bank_churn():
        bank_churn = spark.read.table("bankcatalog.banksilver.silver_bank_churn")
        customerinfo = spark.read.table("bankcatalog.banksilver.silver_customerinfo")    
        result = (        
                  bank_churn        
                  .join(customerinfo, bank_churn.CustomerId == customerinfo.CustomerId, "left")
                  .drop(customerinfo.CustomerId)       
                  .select(*bank_churn.columns, customerinfo.Surname)    
                )    
        return result
 
```
