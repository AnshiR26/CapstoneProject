```sql
USE CATALOG bankcatalog;
USE SCHEMA bankbronze;
```

```sql
CREATE OR REFRESH STREAMING TABLE bronze_active_customers
COMMENT 'Raw active customers data ingested'
TBLPROPERTIES ('quality' = 'bronze')
AS
SELECT *
FROM cloud_files(
  '/Volumes/bankcatalog/bankbronze/bank_data_volume/ActiveCustomer.csv',
  'csv',
  map('cloudFiles.inferColumnTypes', 'true')
);
```
